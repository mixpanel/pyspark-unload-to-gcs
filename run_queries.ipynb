```python
import json

queries_json = dbutils.widgets.get("queries")
queries = json.loads(queries_json)
results = []

for idx, query in enumerate(queries):
    try:
        df = spark.sql(query)
        
        columns = df.columns
        rows = df.collect()
        result_data = [
            [str(val) if val is not None else None for val in row]
            for row in rows
        ]
        
        results.append({
            "columns": columns,
            "result": result_data,
            "error": None
        })
        
    except Exception as e:
        results.append({
            "columns": None,
            "result": None,
            "error": f"Query {idx + 1}: {str(e)}"
        })

result_json = json.dumps(results)
dbutils.notebook.exit(result_json)
```
